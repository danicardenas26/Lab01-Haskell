```haskell
-- ¿Qué significa que Haskell sea un lenguaje fuertemente tipado y evaluado perezosamente?
-- Ser fuertemente tipado implica que el sistema de tipos es rígido y evita que se mezclen 
-- datos incompatibles, detectando errores antes de correr el código. La evaluación 
-- perezosa significa que Haskell posterga el cálculo de las expresiones hasta que el 
-- resultado es realmente indispensable.
```
