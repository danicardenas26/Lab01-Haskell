```haskell
head [10,20,30,40]
tail [10,20,30,40]
take 3 [1..10]
drop 4 [1..10]
length [1,2,3,4,5]
```

```haskell
-¿Qué hace la expresión [1..10] y cómo se relaciona con la evaluación perezosa?
-- La expresión genera una lista aritmética del 1 al 10. Se relaciona con la evaluación 
-- perezosa porque Haskell no genera todos los elementos en memoria de inmediato; 
-- solo calcula cada número conforme las funciones (como take o drop) los van solicitando.
```

```haskell
primerElemento :: [a] -> a
primerElemento (x:_) = x
primerElemento (x:xs) = x
primerElemento [] = error "Lista vacia"
primerElemento [1, 2, 3]
```

```haskell
productoLista :: [Int] -> Int
productoLista [] = 1
productoLista (x:xs) = x * productoLista xs
productoLista[1, 2, 3, 4]
```

```haskell
eliminarNegativos :: [Int] -> [Int]
eliminarNegativos [] = []
eliminarNegativos (x:xs)
    | x >= 0    = x : eliminarNegativos xs
    | otherwise = eliminarNegativos xs
```

```haskell
eliminarNegativos [-1, 2, -3, 5]
```

```haskell
invertir :: [a] -> [a]
invertir [] = []
invertir (x:xs) = invertir xs ++ [x]
```

```haskell
invertir [1, 2, 3]
```
