```python
incrementarLista xs = map (\x -> x + 1) xs
```

```python
incrementarLista[2, 4, 5]
```

```python
filtrarPares = filter (\x -> x `mod` 2 == 0)
```

```python
filtrarPares[1, 2, 3, 4, 5, 6]
```

```python
sumarTuplas xs = map (\(a, b) -> a + b) xs
```

```python
sumarTuplas [(5, 8), (7, 9)]
```
